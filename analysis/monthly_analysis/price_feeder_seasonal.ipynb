## Seasonal Futures Price Feeder Cattle

- Analysing seasonal patterns of Feeder Cattle price by referring to the futures contract's price, since there is more data available for this as compared to the cash price time series from USDA - the two have an extremely strong corelation therefore the below analysis shouldn't be distorted too much (check out this notebook for the [correlation](./futures_cash_price.ipynb))

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.append('../')
import src

import warnings
warnings.filterwarnings('ignore')
```

```python
feeder_cattle_futures = pd.read_csv('../csv/feeder_cattle_futures.csv')
feeder_cattle_futures = src.clean_tv_data(feeder_cattle_futures)

feeder_cattle_futures_adj = src.adjust_series_cpi(feeder_cattle_futures)

feeder_cattle_futures.plot(title='Feeder Cattle Futures ($/cwt)')
```

```python
feeder_cattle_futures_adj.groupby(feeder_cattle_futures_adj.index.month).mean().plot()
plt.title('Mean Price Received Feeder Cattle Futures by Month (Adj. CPI)')
```

- Seasonal pattern of Feeder cattle futures is nearly opposite to that of Live cattle futures (check this [notebook](./price_live_seasonal.ipynb)). 
- Lowest point in March, with peak in July. For live cattle, we had a peak in May, and a low in October.
