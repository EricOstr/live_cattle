# Macro Indicators

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import src

import warnings
warnings.filterwarnings('ignore')
```

```python
gdp_qrt = src.get_us_gdp_qrt_series()

fig, ax = plt.subplots()
sns.lineplot(data=gdp_qrt, ax=ax)

for i, change in enumerate(gdp_qrt.pct_change()):
    if change < 0:
        ax.axvspan(gdp_qrt.index[i-1], gdp_qrt.index[i], color='grey', alpha=0.2)

ax.set_title('US GDP Quarterly Series')
```
