```python
import urllib.request
import pandas as pd
import urllib.parse
from requests.utils import requote_uri
from dotenv import load_dotenv
import os
import io


class usda_quick_stats_dataframe:

    def __init__(self):

        load_dotenv()

        self.base_url_api_get = 'http://quickstats.nass.usda.gov/api/api_GET/?key=' + \
            os.environ['USDA_API_KEY'] + '&'
        self.output_file_path = r'csv_raw/'



    def query(self):
        pass

    def get_data_df(self, parameters):

        s_result = urllib.request.urlopen(self.base_url_api_get + parameters)
        s_text = s_result.read().decode('utf-8')

        df = pd.read_csv(io.StringIO(s_text))

        return df

    def get_data_csv(self, parameters, file_name):

        s_result = urllib.request.urlopen(self.base_url_api_get + parameters)
        s_text = s_result.read().decode('utf-8')

        s_file_name = self.output_file_path + file_name + ".csv"
        o_file = open(s_file_name, "w", encoding="utf8")
        o_file.write(s_text)

        o_file()


    def create_query_string(self,
                            source_desc='SURVEY',
                            sector_desc='FARMS & LANDS & ASSETS',
                            commodity_desc='FARM OPERATIONS',
                            statisticcat_desc='AREA OPERATED',
                            unit_desc='ACRES',
                            freq_desc='ANNUAL',
                            reference_period_desc='YEAR',
                            year__GE='2010',
                            agg_level_desc='NATIONAL',
                            state_name='US TOTAL',
                            format=None,
                            ):

        return f'source_desc={source_desc}' +  \
            '&' + urllib.parse.quote(f'sector_desc={sector_desc}') + \
            '&' + urllib.parse.quote(f'commodity_desc={commodity_desc}') + \
            '&' + urllib.parse.quote(f'statisticcat_desc={statisticcat_desc}') + \
            f'&unit_desc={unit_desc}' + \
            f'&freq_desc={freq_desc}' + \
            f'&reference_period_desc={reference_period_desc}' + \
            f'&year__GE={year__GE}' + \
            f'&agg_level_desc={agg_level_desc}' + \
            '&' + urllib.parse.quote('state_name=US TOTAL') + \
            '&format=CSV'


        # return f'source_desc={source_desc}' +  \
        #     f'&' + urllib.parse.quote(f'sector_desc={sector_desc}') + \
        #     f'&' + urllib.parse.quote(f'commodity_desc={commodity_desc}') + \
        #     f'&' + urllib.parse.quote(f'statisticcat_desc={statisticcat_desc}') + \
        #     f'&' + urllib.parse.quote(f'unit_desc={unit_desc}') + \
        #     f'&freq_desc={freq_desc}' + \
        #     f'&reference_period_desc={reference_period_desc}' + \
        #     f'&year__GE={year__GE}' + \
        #     f'&agg_level_desc={agg_level_desc}' + \
        #     '&' + urllib.parse.quote(f'state_name={state_name}') + \
        #     f'&format={format}'
```

```python
client = usda_quick_stats_dataframe()
param = client.create_query_string(
    source_desc='SURVEY',
    sector_desc='LIVESTOCK',
    commodity_desc='BEEF',
    statisticcat_desc='PRODUCTION',
    unit_desc='LB',
    freq_desc='MONTHLY',
    reference_period_desc='YEAR',
    year__GE='2020',
    agg_level_desc='NATIONAL',
    state_name='US TOTAL',
    format=None,
)
df = client.get_data_df(param)
df
```
